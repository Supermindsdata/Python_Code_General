#### **Helper Functions**

```python
def missing_values(df):
    percent_missing = df.isnull().mean()*100
    missing_value_df = pd.DataFrame(percent_missing, columns=['% Missing Values'])
    missing_value_df.sort_values(by='% Missing Values', ascending=False, inplace=True)

    return missing_value_df
```
